Vamos trabalhar com um famoso conjunto de dados do titanic para estes exercícios:

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python

titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

## Exercícios
** Recrie os gráficos abaixo usando o dataframe titanic. Existem muito poucas dicas, pois a maioria dos gráficos pode ser feita com apenas uma ou duas linhas de código e uma dica basicamente revelaria a solução. Preste muita atenção aos rótulos x e y para obter dicas.**

* *Observação! Para não perder a imagem do gráfico, certifique-se de não codificar na célula que está diretamente acima do gráfico, há uma célula extra acima daquela que não substituirá o gráfico! **

```python
scatter_plot = sns.jointplot(x='fare', y='age', data=titanic, kind='scatter')
 
plt.show()

```

```python
sns.histplot(titanic['fare'],bins=30,kde=False,color='r')

plt.show()
```

```python
box_plot = sns.boxplot(x='class', y='age', data=titanic, palette='rainbow')
 
plt.show()
```

```python
sns.swarmplot(x='class',y='age',data=titanic,palette='Set2')

plt.show()
```

```python
count_plot = sns.countplot(x='sex', data=titanic)
plt.show()
```

```python

numero_colunas = titanic.select_dtypes(include=['number'])
heatmap = sns.heatmap(numero_colunas.corr(), annot=False, cmap='coolwarm')
heatmap.set_title('Titanic.corr()')
plt.yticks(rotation=0)
plt.xticks(rotation=90) # fazer rotacao das legendas


plt.show()
```

```python

g = sns.FacetGrid(data=titanic,col='sex')
g.map(plt.hist,'age')
g.set(ylim=(0, 120))

plt.show()
```
